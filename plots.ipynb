```python
import os
import sys
import numpy as np
import torch
import matplotlib.pyplot as plt
import seaborn as sns


def flater(path, idx):
    input = torch.load(path)
    evolution = []
    output = []
    for i in input:
        output.append(torch.cat(i))
    
    for i in output:
        evolution.append(i[idx].tolist())

    #Make all the elements in evolution a string
    evolution = [str(i) for i in evolution]
    foo[0] = 1
    return evolution


foo  = flater("outputs0.pt",92)

print(foo)


foo[0] = '1'
# Define the desired order of categories
desired_order = [str(i) for i in sorted(set(foo), key=lambda x: int(x) if x.isdigit() else float('inf'))]

# Convert categories to numerical values
numerical_foo = [desired_order.index(val) for val in foo]

plt.figure(figsize=(12, 4))
plt.scatter(range(len(foo)), numerical_foo, s=80)

# Set y-axis tick labels to the original categorical values
plt.yticks(range(len(desired_order)), desired_order)

plt.xlabel('Epoch')
plt.ylabel('Label')
plt.title('Label evolution during poisoned training, O-T-O')
```
